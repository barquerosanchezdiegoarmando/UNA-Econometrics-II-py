```python
# Creación de un Boxplot e Histograma con la misma escala .


def histogram_boxplot(data, feature, figsize=(15, 10), kde=False, bins=None):
    """
    Boxplot e Histogram combinados

    data: dataframe
    feature: dataframe column
    figsize: size of figure (default (15,10))
    kde: whether to show the density curve (default False)
    bins: number of bins for histogram (default None)
    """
    f2, (ax_box2, ax_hist2) = plt.subplots(
        nrows=2,  # Number of rows of the subplot grid= 2
        sharex=True,  # x-axis will be shared among all subplots
        gridspec_kw={"height_ratios": (0.25, 0.75)},
        figsize=figsize,
    )  # creating the 2 subplots
    sns.boxplot(
        data=data, x=feature, ax=ax_box2, showmeans=True, color="violet"
    )  # boxplot will be created and a triangle will indicate the mean value of the column
    sns.histplot(
        data=data, x=feature, kde=kde, ax=ax_hist2, bins=bins
    ) if bins else sns.histplot(
        data=data, x=feature, kde=kde, ax=ax_hist2
    )  # For histogram
    ax_hist2.axvline(
        data[feature].mean(), color="green", linestyle="--"
    )  # Add mean to the histogram
    ax_hist2.axvline(
        data[feature].median(), color="black", linestyle="-"
    )  # Add median to the histogram
```

```python
histogram_boxplot(datos, 'ExpFOB')
```

### *Librerías específicas para realizar el pronóstico*

```python
# librerias especificas para esto 
from statsmodels.tsa.api import SimpleExpSmoothing
import warnings
from statsmodels.tools.sm_exceptions import ConvergenceWarning
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore', ConvergenceWarning)
```

```python
# tiramos el modelo 
# optimize = True lo que hace es optimizar los coeficientes alpha, beta y gamma
modelo = ExponentialSmoothing(datos['ExpFOB'], seasonal_periods = 12, trend='mul', seasonal='mul').fit(optimized = True)
```

```python
# veamos que valores utilizo 
modelo.summary()
```

# Alpha = 0.5886465
# Beta = 0.0357000
# Gamma = 0.3532970

## *Asignamos los valores estimados al dataframe original*

```python

datos['Holt_Winters'] = modelo.fittedvalues
```

```python
datos
```

## *Realizamos la proyección de datos a t+3*

```python
exp_f = modelo.forecast(3)

exp_f
```

### *Realizamos un gráfico que posee los observados, estimados y los proyectados*

```python

# voy a crear un grafico donde se observe la predicción en otro color 
plt.figure(figsize = (20,7))
plt.plot(datos.index, datos['ExpFOB'], '-o', color = 'black', label = 'Yi')
plt.plot(datos.index, datos['Holt_Winters'], '-o', color = 'blue', label = 'Fi')
plt.plot(exp_f, '-o', color = 'green', label = 'Fi+1')
plt.xlabel('Periodos')
plt.ylabel('Exportaciones')
plt.title('Serie de tiempo')
plt.legend(loc='best')
plt.show()
```
